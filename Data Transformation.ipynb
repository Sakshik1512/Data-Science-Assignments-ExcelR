```python
df.replace("?", np.nan, inplace=True)

# Check again
print(df.isnull().sum())

# Fill categorical with mode
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

# Fill numerical with median
for col in df.select_dtypes(include=np.number):
    df[col].fillna(df[col].median(), inplace=True)
```

Explanation:

Mode → categorical
Median → numerical (robust to outliers)

```python
from sklearn.preprocessing import StandardScaler

num_cols = df.select_dtypes(include=np.number).columns

scaler_std = StandardScaler()
df_std = df.copy()
df_std[num_cols] = scaler_std.fit_transform(df[num_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_mm = MinMaxScaler()
df_mm = df.copy()
df_mm[num_cols] = scaler_mm.fit_transform(df[num_cols])
```

```python
low_card_cols = [col for col in df.columns if df[col].nunique() < 5 and df[col].dtype == 'object']

df = pd.get_dummies(df, columns=low_card_cols, drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

high_card_cols = [col for col in df.columns if df[col].nunique() >= 5 and df[col].dtype == 'object']

le = LabelEncoder()
for col in high_card_cols:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(contamination=0.01)
outliers = iso.fit_predict(df.select_dtypes(include=np.number))

df = df[outliers == 1]
```
