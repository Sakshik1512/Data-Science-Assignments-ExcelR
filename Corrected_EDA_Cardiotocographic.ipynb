# Exploratory Data Analysis (EDA) on Cardiotocographic Dataset


## Objective
The goal of this analysis is to perform a complete Exploratory Data Analysis (EDA) on the Cardiotocographic dataset.

This notebook includes:
- Data cleaning and preparation
- Missing value handling
- Correct outlier detection and treatment
- Statistical summaries
- Histograms, boxplots, heatmaps
- Pair plots and violin plots
- Pattern recognition and insights
- Final conclusion


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10,6)

# Load dataset
df = pd.read_csv('Cardiotocographic.csv')

# Display first rows
df.head()
```

## Dataset Overview

```

# Shape of dataset
print("Dataset Shape:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Data types
print("\nData Types:")
print(df.dtypes)

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

## Data Cleaning and Preparation

```

# Check duplicates
print("Duplicate Rows:", df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()

# Convert numeric columns if needed
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing values using median
df = df.fillna(df.median(numeric_only=True))

print("Missing values after treatment:")
print(df.isnull().sum())
```

## Correct Outlier Detection and Treatment

```

# Function to detect outliers using IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

    return lower_bound, upper_bound, outliers

# Detect outliers for all numerical columns
outlier_summary = {}

for col in df.columns:
    lb, ub, outliers = detect_outliers_iqr(df, col)
    outlier_summary[col] = len(outliers)

# Display number of outliers
print("Outlier Summary:")
for key, value in outlier_summary.items():
    print(f"{key}: {value}")
```

```

# Treat outliers using capping method
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print("Outlier treatment completed.")
```

## Statistical Summary

```

# Descriptive statistics
summary = df.describe().T

# Add IQR
summary['IQR'] = summary['75%'] - summary['25%']

summary[['mean', '50%', 'std', 'IQR', 'min', 'max']]
```
