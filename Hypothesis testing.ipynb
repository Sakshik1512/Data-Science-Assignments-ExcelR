# Confidence Interval

# 1. Import Libraries

```python
import numpy as np
from scipy import stats
```

# 2. Input Data

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 
        1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
data
```

```python
n = len(data)
n
```

# 3. Calculate Mean and Standard Deviation

```python
mean = np.mean(data)
std = np.std(data, ddof=1)  # sample standard deviation

print("Sample Size (n):", n)
print("Sample Mean:", round(mean, 4))
print("Sample Standard Deviation:", round(std, 4))
```

# 4. 99% Confidence Interval (t-distribution)

```python
# t critical value
t_critical = stats.t.ppf(0.995, df=n-1)

# margin of error
margin_t = t_critical * (std / np.sqrt(n))

# confidence interval
ci_t_lower = mean - margin_t
ci_t_upper = mean + margin_t

print("\n99% CI using t-distribution:")
print("Lower Limit:", round(ci_t_lower, 4))
print("Upper Limit:", round(ci_t_upper, 4))
```

# 5. 99% Confidence Interval (Z-distribution)

```python
sigma = 0.2  # population standard deviation
z_critical = 2.576  # for 99% CI

margin_z = z_critical * (sigma / np.sqrt(n))

ci_z_lower = mean - margin_z
ci_z_upper = mean + margin_z

print("\n99% CI using Z-distribution:")
print("Lower Limit:", round(ci_z_lower, 4))
print("Upper Limit:", round(ci_z_upper, 4))
```

# 6. Final Summary

```python
print("\n--- Final Summary ---")
print("Sample Mean:", round(mean, 4))

print("\nUsing t-distribution (unknown population std):")
print(f"99% CI = ({round(ci_t_lower,4)}, {round(ci_t_upper,4)})")

print("\nUsing Z-distribution (known population std = 0.2):")
print(f"99% CI = ({round(ci_z_lower,4)}, {round(ci_z_upper,4)})")
```

# Hypothsis Testing

```python
# Step 1: Import library
import numpy as np
from scipy.stats import norm
```

```python
# Step 2: Given values
x_bar = 3050   # sample mean
x_bar
X = 600
X
mu = 1000 + 5*X   # theoretical mean
mu
sigma = 5 * 25    # population std
sigma
n = 25
n
```

```python
# Step 3: Calculate Z-statistic
Z = (x_bar - mu) / (sigma / np.sqrt(n))
Z
print("Theoretical Mean (mu):", mu)
print("Z-statistic:", Z)
```

```python
# Step 4: Critical value
alpha = 0.05
z_critical = norm.ppf(1 - alpha)

print("Critical Value (Z):", z_critical)
```

```python
# Step 5: Decision
if Z > z_critical:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# Step 6: Conclusion
print("\nConclusion:")
print("There is no evidence that weekly operating costs are higher than expected.")
```
