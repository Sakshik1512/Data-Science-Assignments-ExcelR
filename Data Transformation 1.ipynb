```python
# ============================================================
# DATA PREPROCESSING AND FEATURE ENGINEERING IN MACHINE LEARNING
# Adult Income Dataset
# ============================================================

# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# ============================================================
# 2. LOAD DATASET
# ============================================================

df = pd.read_csv("adult_with_headers (1).csv")

# Display first 5 rows
print("First 5 Rows:")
print(df.head())
```

```python
# ============================================================
# 3. BASIC DATA EXPLORATION
# ============================================================

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nSummary Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ============================================================
# 4. HANDLE MISSING VALUES
# ============================================================

# Replace '?' with NaN if present
df.replace(' ?', np.nan, inplace=True)
df.replace('?', np.nan, inplace=True)

print("\nMissing Values After Replacing ?:")
print(df.isnull().sum())

# Separate Numerical and Categorical Columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute Numerical Columns using Mean
num_imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = num_imputer.fit_transform(df[numerical_cols])

# Impute Categorical Columns using Most Frequent
cat_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])

print("\nMissing Values After Imputation:")
print(df.isnull().sum())
```

```python
# ============================================================
# 5. SCALING TECHNIQUES
# ============================================================

# Copy DataFrame
scaled_df = df.copy()

# ------------------------------------------------------------
# Standard Scaling
# ------------------------------------------------------------

standard_scaler = StandardScaler()

scaled_df[numerical_cols] = standard_scaler.fit_transform(
    scaled_df[numerical_cols]
)

print("\nStandard Scaling Applied")

# ------------------------------------------------------------
# Min-Max Scaling
# ------------------------------------------------------------

minmax_df = df.copy()

minmax_scaler = MinMaxScaler()

minmax_df[numerical_cols] = minmax_scaler.fit_transform(
    minmax_df[numerical_cols]
)

print("\nMin-Max Scaling Applied")
```

```python
# ============================================================
# SCALING DISCUSSION
# ============================================================

print("""

SCALING TECHNIQUES DISCUSSION

1. Standard Scaling:
   - Converts data into mean = 0 and standard deviation = 1
   - Preferred when data follows normal distribution
   - Useful for algorithms like Logistic Regression, SVM, PCA

2. Min-Max Scaling:
   - Scales values between 0 and 1
   - Preferred when exact boundaries are needed
   - Useful for Neural Networks and KNN

""")
```

```python
# ============================================================
# 6. ENCODING TECHNIQUES
# ============================================================

encoded_df = df.copy()

# ------------------------------------------------------------
# One-Hot Encoding
# Apply to categorical columns with less than 5 categories
# ------------------------------------------------------------

small_cat_cols = []

for col in categorical_cols:
    if encoded_df[col].nunique() < 5:
        small_cat_cols.append(col)

print("\nColumns for One-Hot Encoding:")
print(small_cat_cols)

encoded_df = pd.get_dummies(encoded_df,
                            columns=small_cat_cols,
                            drop_first=True)
```

```python
# ------------------------------------------------------------
# Label Encoding
# Apply to categorical columns with more than 5 categories
# ------------------------------------------------------------

label_encoder = LabelEncoder()

large_cat_cols = []

for col in categorical_cols:
    if df[col].nunique() >= 5:
        large_cat_cols.append(col)

print("\nColumns for Label Encoding:")
print(large_cat_cols)

for col in large_cat_cols:
    encoded_df[col] = label_encoder.fit_transform(encoded_df[col])

print("\nEncoding Completed")
```

```python
# ============================================================
# ENCODING DISCUSSION
# ============================================================

print("""

ENCODING TECHNIQUES DISCUSSION

1. One-Hot Encoding:
   Pros:
   - No ordinal relationship introduced
   - Works well for nominal categories

   Cons:
   - Increases dimensionality
   - Can create sparse data

2. Label Encoding:
   Pros:
   - Simple and memory efficient
   - Faster computation

   Cons:
   - Introduces ordinal relationship
   - Can mislead some ML algorithms

""")
```

```python
# ============================================================
# 7. FEATURE ENGINEERING
# ============================================================

feature_df = encoded_df.copy()

# ------------------------------------------------------------
# Feature 1: Capital Difference
# ------------------------------------------------------------

# Feature Engineering
feature_df['capital_difference'] = (
    df['capital_gain'] - df['capital_loss']
)

# Log Transformation
print(df['capital_gain'].skew())

feature_df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

```python
# ============================================================
# FEATURE ENGINEERING JUSTIFICATION
# ============================================================

print("""

FEATURE ENGINEERING JUSTIFICATION

1. capital_difference:
   - Combines capital gain and capital loss information
   - Helps capture overall financial status

2. age_group:
   - Groups ages into meaningful categories
   - Helps models identify age-based patterns easily

""")
```

```python
# ============================================================
# 8. LOG TRANSFORMATION ON SKEWED FEATURE
# ============================================================

print("\nSkewness of Numerical Features:")
print(df[numerical_cols].skew())

# Selecting highly skewed feature
print("\nSkewness Before Transformation:")
print(df['capital-gain'].skew())

# Apply Log Transformation
feature_df['capital_gain_log'] = np.log1p(df['capital-gain'])

print("\nSkewness After Transformation:")
print(feature_df['capital_gain_log'].skew())
```

```python
# ============================================================
# VISUALIZATION BEFORE AND AFTER TRANSFORMATION
# ============================================================

fig, ax = plt.subplots(1, 2, figsize=(12,5))

# Before Transformation
ax[0].hist(df['capital-gain'], bins=30)
ax[0].set_title("Before Log Transformation")
ax[0].set_xlabel("capital-gain")
ax[0].set_ylabel("Frequency")

# After Transformation
ax[1].hist(feature_df['capital_gain_log'], bins=30)
ax[1].set_title("After Log Transformation")
ax[1].set_xlabel("capital_gain_log")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# LOG TRANSFORMATION JUSTIFICATION
# ============================================================

print("""

LOG TRANSFORMATION JUSTIFICATION

The feature 'capital-gain' was selected because it is highly positively skewed.
Most observations contain small or zero values, while a few observations
contain very large values.

Applying log transformation helps:
1. Reduce skewness
2. Compress extreme values
3. Improve feature distribution
4. Enhance machine learning model performance

After transformation, the feature becomes more balanced
and suitable for machine learning algorithms.

""")
```
